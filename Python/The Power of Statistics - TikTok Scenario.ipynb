## Hypothesis testing

- **H_0$**: There is no difference in number of views between tiktok videos posted by verified and unverified accounts
- **H_A$**: There is a difference in number of views. 

We choose 5% as the significance level and proceed with a two-sample t-test.

```python
not_verified = data[data["verified_status"] == "not verified"]["video_view_count"]
verified = data[data["verified_status"] == "verified"]["video_view_count"]

# Implement a t-test using the two samples
stats.ttest_ind(a=not_verified, b=verified, equal_var=False)
```

**p-value = 2.61e-120**

The p-value is much smaller that the significance level of 5%. As a result, we can reject the null hypothesis and say that the difference in views in statistically significant.
